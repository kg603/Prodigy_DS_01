## Distribution of Male and Female in a population

```python
years = [str(year) for year in range(1960, 2024)]  # Adjusting the range based  dataset
total_male_population = Male[years].sum()
total_female_population = Female[years].sum()
total_population = total_male_population + total_female_population

# Converting to DataFrame for visualization
population_df = pd.DataFrame({
    'Year': years,
    'Male': total_male_population.values,
    'Female': total_female_population.values,

})
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'population_df' is a DataFrame with columns 'Year', 'Male', and 'Female'

# Positions for the bars
x = np.arange(len(population_df['Year']))

# Bar width for side-by-side bars
bar_width = 0.4

# Create the bar chart
plt.figure(figsize=(14, 7))

# Male bars shifted slightly left
plt.bar(x - bar_width / 2, population_df['Male'], width=bar_width, color='blue', edgecolor='black', label='Male')

# Female bars shifted slightly right
plt.bar(x + bar_width / 2, population_df['Female'], width=bar_width, color='red', edgecolor='black', label='Female')

# Add labels and title
plt.title('Total Population since 1960', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(x, population_df['Year'], rotation=90, fontsize=10)
plt.legend()
plt.grid(axis='y', alpha=0.4)

# Show the chart
plt.tight_layout()
plt.show()
```
